Plot discontinuity threshold noise interaction

```python
sns.set(style="white")
pal = sns.color_palette("bright")

size=6
fig, (ax1) = plt.subplots(1,1, sharey=False, sharex=True, figsize=[size, size/1.618])

    # temperature

tempvalues = df_scen_temp["scndc"].values
years = df_scen_temp["year"].values

fit = np.polyfit(years, tempvalues, 5)
f = np.poly1d(fit)
novartemp = f(range(years[0],years[-1]+1))

threshold = 3.2

ax1.plot(df_scen_temp["year"], np.zeros(len(novartemp))+threshold, dashes=[6, 2], linewidth=3, color='black', label = 'Threshold')
ax1.plot(df_scen_temp["year"], df_scen_temp["scndc"], linewidth=3, color=pal[2], label = 'Temperature \n with variability')
ax1.plot(df_scen_temp["year"], novartemp, linewidth=3, color=pal[0], label = 'Climatic Temperature')
ax1.set_ylabel('Global temperature')
plt.xlabel('Years')
plt.legend(loc="best")
plt.xlim([2015,2160])
fig.savefig('discontinuityNoise.jpg', dpi = 500)
```

### resulting burke damages: follow calculation in PAGE-ICE

```python
realisedanomalyburke = [1.2016316666666669, 1.5611749999999998, 1.2678633333333336, 1.107015, 0.9556283333333334, 1.1448616666666667, 0.9840133333333334, 1.1543233333333334]
def calc_burke(temp, r):
    burkeregionalimpact = temp - realisedanomalyburke[r] 
    i1log = (-0.00829990966469437  * (burkeregionalimpact - baseline[r])) + (-0.000500003403703578 * ((burkeregionalimpact - 21.)**2 - (baseline[r] - 21.)**2))
    impact = 100 * (1-np.exp(i1log))
    
    return impact

             
df_damages = pd.DataFrame()   

for r in range(8):
    randtemps = np.random.normal(baseline[r], stddev[r], numdraws)
    for rand in range(numdraws):
        damage = calc_burke(randtemps[rand], r)
        df_damages = df_damages.append({'Region': region[r], 'Damage': damage}, ignore_index=True)
             
```

Find the GDP-weighted impact

```python
GDP = [1.7276714501653694e7, 
       4.163916626016402e6, 
       1.6755755949456695e7, 
       1.5042627818163713e7, 
       1.1780218688261595e7, 
       7.107101842851756e6, 
       7.712450726531674e6, 
       9.422720364817122e6]
realisedanomalyburke = [1.2016316666666669, 1.5611749999999998, 1.2678633333333336, 1.107015, 0.9556283333333334, 1.1448616666666667, 0.9840133333333334, 1.1543233333333334]
todollar = 1000000
def calc_burke_gdp(temp, r):
    burkeregionalimpact = temp - realisedanomalyburke[r] 
    i1log = (-0.00829990966469437  * (burkeregionalimpact - baseline[r])) + (-0.000500003403703578 * ((burkeregionalimpact - 21.)**2 - (baseline[r] - 21.)**2))
    impact = 100 * (1-np.exp(i1log))
    impact = impact * GDP[r]
    
    return impact*todollar

             
df_damages_gdp = pd.DataFrame()   

for r in range(8):
    randtemps = np.random.normal(baseline[r], stddev[r], numdraws)
    for rand in range(numdraws):
        damage_gdp = calc_burke_gdp(randtemps[rand], r)
        df_damages_gdp = df_damages_gdp.append({'Region': region[r], 'Damage_gdp': damage_gdp}, ignore_index=True)
```

## Comparing different scenarios

#### Social Cost of Carbon - with uncertainty in only the var_multiplier parameters
Reading in the results from montecarlo outputs

```python
# scenarios I would also like to do: ["1_5C", "2_0C",]
scenarios = ["2_5C", "NDC", "BAU", "RCP2_6_SSP1", "RCP4_5_SSP2", "RCP8_5_SSP5"]
mcruns = 500
scc_sc = np.zeros((len(scenarios),mcruns)) # 8 scenarios, 100 mc runs per scenario
i = 0
for sc in scenarios:
    
    path = "mcPAGEVAR\onlyvarMC\%s\scc.csv" % (sc)
    data = genfromtxt(path, delimiter=',')
    scc_sc[i] = data
    
    i = i + 1
```

```python
MCruns = mcruns
pos = range(len(scenarios))
fig2, axs = plt.subplots()
fig2.set_size_inches(8, 8)
# plt.yscale('log')
axs.violinplot(np.transpose(scc_sc), pos, points=20, widths=0.5,
                      showmeans=True, showextrema=True, showmedians=False)
axs.set_title('Social Cost of Carbon with only var_multiplier in MCS (line = mean), MC = '+ str(MCruns))
axs.set_xticks(range(len(scenarios)))
axs.set_xticklabels(scenarios)

df = pd.DataFrame(columns=['Scenario', 'SCC mean', 'SCC median', 'SCC min', 'SCC max'])

for i in range(len(scenarios)):
    df.loc[i] = (scenarios[i], np.mean(scc_sc[i]), np.median(scc_sc[i]), np.min(scc_sc[i]), np.max(scc_sc[i]))
    
df.head()
```
